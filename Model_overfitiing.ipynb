Consider the dataset generated as follows:

a) The datset is 2D - has two columns.Has 1500 labelled instances and each of the instances is assigned 1 or 0

e.g <br>
2.3 0.8 0 <br>
2.5 9.9 1<br>


....
1500 4.5 4. 0

Generate data using the following approaches:

1. For class 1 are generated from a mixture of 3 gaussian distribution - normal distribution.
The normal distribution will be centered at [6,14] , [10,6], [14,14]

2. For class 0 are generated using uniform distribution in a square region

3. Equal number of labelled instances


```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random
```

```
%matplotlib inline
```

```
N = 1500
mean1 = [6,14]
mean2 = [10,6]
mean3 = [14,14]
cov = [[3.5,0], [0,3.5]]
cov
```

```
# generate the data
np.random.seed(50)

x = np.random.multivariate_normal(mean1, cov, int(N/6))
```

```
x[:,0]
```

```
np.average(x[:,0])
```

```
len(x)
```

```
x = np.concatenate((x, np.random.multivariate_normal(mean2, cov, int(N/6))))
```

```
len(x)
```

```
x = np.concatenate((x, np.random.multivariate_normal(mean3, cov, int(N/6))))
len(x)
```

```
x = np.concatenate((x, 20*np.random.rand(int(N/2),2)))
len(x)
```

```
y = np.concatenate((np.ones(int(N/2)), np.zeros(int(N/2))))
y
```

```
plt.plot(x[:int(N/2), 0], x[:int(N/2), 1], 'r+', x[int(N/2):, 0], x[int(N/2):,1], 'k.', ms=4)
```

```
from sklearn.model_selection import train_test_split
# random state for reproducability
# training 80%
x_train, x_test, y_train, y_test =  train_test_split(x,y, test_size = 0.2, random_state=1)
x_train
```

```
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```
# as depth increases, becomes more complex, becomes more generalised
maxDepths = [2,3,4,5,6,7,8,9,10,15,20,25,30,35,40,45,50]
```

```
# when the depth is 2 what is the model accuracy on training test and testset

trainAcc = np.zeros(len(maxDepths))
testAcc = np.zeros(len(maxDepths))
testAcc
```

```
index = 0

for depth in maxDepths:
  clf = tree.DecisionTreeClassifier(max_depth=depth)
  clf = clf.fit(x_train, y_train)
  y_predicted_train = clf.predict(x_train)
  y_predicted_test = clf.predict(x_test)
  trainAcc[index] = accuracy_score(y_train, y_predicted_train)
  testAcc[index]= accuracy_score(y_test, y_predicted_test)
  index +=1
  
```

```
trainAcc
```

```
testAcc
```

```
plt.plot(maxDepths, trainAcc, 'ro-', maxDepths, testAcc, 'bv--')
plt.legend(['Training Accuracy', 'Test Accuracy'])
plt.xlabel('Max depth')
plt.ylabel('Accuracy')
```

The plot shows that the accuracy will continue to improve as the max depth becomes more accurate and the test accuracy improves up tp 5 but decreases after due to model overfiting
